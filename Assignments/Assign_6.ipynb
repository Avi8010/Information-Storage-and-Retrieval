```python
import matplotlib.image as img
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
import numpy as np
import pandas as pd
```

```python
image=img.imread("Joker.jpg")
plt.imshow(image)
```

```python
red, yellow=image.copy(),image.copy()
```

```python
red[:,:,(1,2)]=0
yellow[:,:,2]=0
```

```python
plt.imshow(red)
```

```python
plt.imshow(yellow)
```

```python
gray_image=rgb2gray(image)
```

```python
plt.imshow(gray_image)
```

```python
print(image.shape)
print(gray_image.shape)
```

```python
thresh=threshold_otsu(gray_image)
binary=gray_image>thresh
plt.imshow(binary)
```

```python
from skimage.filters import gaussian
blurred_image = gaussian(image,sigma=1)
plt.imshow(blurred_image)
```

```python
plt.hist(gray_image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

```python
feature_values = np.array(gray_image)

# Create feature headers (column names)
feature_headers = [f"Pixel {i}" for i in range(len(feature_values))]

# Store the feature headers and values in a DataFrame
data = pd.DataFrame({header: [value] for header, value in zip(feature_headers, feature_values)})

# Save the data to a CSV file
data.to_csv('image_features.csv', index=True)
data.head()
```
